```python
import os
import sys

from IPython.display import IFrame
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
# import missingno as msno
# import seaborn as sns
# import matplotlib.patches as patches
# import warnings
# import networkx as nx
# import osmnx as ox
# import requests
# import matplotlib.cm as cm
# import matplotlib.colors as colors

import urllib
import zipfile
import glob
```

## Downloading South Korea geodata through Natural Earth

```python
# from IPython.display import IFrame

IFrame(src='https://www.naturalearthdata.com', width=1000, height=500)
```

```python
# import urllib

ne_admin1_states_url='https://www.naturalearthdata.com/http//www.naturalearthdata.com/download/10m/cultural/ne_10m_admin_1_states_provinces.zip'
```

## Download the zipfile and store it in the Data folder

```python
urllib.request.urlretrieve(ne_admin1_states_url, "Data/ne_admin_1_states_provinces.zip")
```

## Unzip the zipfile and Extract

```python
# import zipfile

ne_admin1_states_zip = zipfile.ZipFile('Data/ne_admin_1_states_provinces.zip')

ne_admin1_states_zip
```

```python
ne_admin1_states_zip.extractall('Data/ne_admin_1_states_provinces')
```

## Look in the directory and see what files are available

```python
# import glob

glob.glob('Data/ne_admin_1_states_provinces/*')
```

## Read the shapefile in this directory

```python
ne_admin_1_states_gdf = gpd.read_file('Data/ne_admin_1_states_provinces/ne_10m_admin_1_states_provinces.shp')

ne_admin_1_states_gdf.head()
```

```python
ne_admin_1_states_gdf.shape
```

```python
ne_admin_1_states_gdf.info()
```

## Filter out the gdf and get 41 admin='South Korea'

```python
south_korea_gdf=ne_admin_1_states_gdf[ne_admin_1_states_gdf['admin']=='South Korea']

south_korea_gdf
```

## Get the columns you need and make the dataframe easier to read

```python
south_korea_gdf_filtered = south_korea_gdf[
    ['geonunit', 'gu_a3', 'gn_name', 'name_de', 
     'fips', 'latitude', 'longitude', 'geometry', 'wikidataid']]

south_korea_gdf_filtered
```

## Plot South Korea

```python
south_korea_gdf_filtered.plot(figsize=(10,10));
```
